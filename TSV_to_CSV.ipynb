En cuanto al campo account_number, no voy a tomar la decisión de eliminar los duplicados porque no sé a ciencia cierta si es un campo que no se puede duplicar. Podría ser el caso que ese account_number vincule a 2 individuos a una misma cuenta, si pensamos en tarjetas de crédito, podrían ser el titular y un adicional.

```python
# LOS EMAILS NO SON CASE SENSITIVE POR LO QUELOS LLEVAMOS A TODOS A LOWERCASE

df_encoding_correcto['email'] =  df_encoding_correcto['email'].apply(lambda x: x.lower())
df_encoding_correcto['email'] = df_encoding_correcto['email'].str.strip()
```

```python
# ESTE REGEX PUEDE NO CAPTAR TODOS LOS EMAILS INVALIDOS

regex = r"(^[a-z0-9_.+-]+@[a-z0-9-]+\.[a-z0-9-.]+$)"


def mail_valido(mail):   
  
    if(re.search(regex,mail)):   
        return 'Valido'  
    else:   
        return 'Invalido'
```

```python
df_encoding_correcto['mail_valido'] = df_encoding_correcto['email'].apply(lambda x: mail_valido(x))
```

```python
df_encoding_correcto['mail_valido'].value_counts()
```

Los mails parecen ser correctos bajo el regex implementado (puede no ser 100% efectivo)

Pasamos entonces a intentar hacer estos pasos repetibles y reutilizables

```python
# Defino una función reutilizable, a tener en cuenta que también utiliza la funcion "mail_valido"
# Esta función devuelve los datos con una columna adicional que indica si el mail es valido o invalido para revisar

def tsv_to_csv(tsv_file_name, csv_file_name):
    
    df_aux = pd.read_table(tsv_file_name, sep='\t', encoding='UTF-16LE')
    
    df_aux['id'] = pd.to_numeric(df_aux['id'], errors='coerce', downcast='integer')

    df_aux = df_aux.dropna()
    
    df_aux['id'] = df_aux['id'].astype(int)
    
    if len(df_aux['id'].unique()) == len(df_aux):
        
        # Para borrar la menor cantidad de datos nos quedamos con el primero de los duplicados (si hubiera)
        # De todas formas habría que tener más cosas en cuenta como si puedo verificar desde otro lado cual es el correcto
        df_aux = df_aux.drop_duplicates(keep='first', subset='id')
    
    df_aux['email'] =  df_aux['email'].apply(lambda x: x.lower())
    df_aux['email'] = df_aux['email'].str.strip()
    df_aux['mail_valido'] = df_aux['email'].apply(lambda x: mail_valido(x))
    
    df_aux = df_aux.sort_values(by='id').reset_index(drop=True)
    
    df_aux.to_csv(csv_file_name, sep='|', encoding='UTF-8', index=False)
    
    return df_aux
```

```python
# EJEMPLO DE UTILIZACIÓN

df = tsv_to_csv('datos_data_engineer.tsv', 'datos_data_engineer.csv')
```
